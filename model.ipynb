1. Problem Statement and Goal

The goal of this task is to perform Exploratory Data Analysis (EDA) on the Iris dataset. 
We aim to understand the relationships between physical measurements of different iris flower
species and identify patterns or outliers using statistical summaries and visualizati

```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load Data set

```python
df=sns.load_dataset('iris');
```

Shape, Column Name , and First few Rows

```python
print("\n Shape of Data Set : \t", df.shape);
print("\n Colums of Data Set : \t", df.columns);
print("\n Rows of Data Set : \n", df.head(5));
```

Data Info and Describe 

```python
print("\n--- Data Info ---")
df.info() 

print("\n--- Summary Statistics ---")
print(df.describe())
```

‚óè	Visualize the dataset:

Relationship between Features

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species')
plt.title('Petal Length vs Petal Width')
plt.show()
```

Histograms
Histograms show us the frequency of specific measurements

```python
df.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()
```

Box Plots


```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Box Plot of Features')
plt.show()
```
